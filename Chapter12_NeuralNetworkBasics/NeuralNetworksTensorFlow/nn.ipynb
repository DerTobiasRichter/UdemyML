```python
from keras.models import *
from keras.layers.core import *
from keras.optimizers import *
import numpy as np
```

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
```

```python
model = Sequential() # 2 Input Neuronen
model.add(Dense(8, input_dim=2)) # 8 Neuronen im n√§chsten Layer, (Hidden Layer 1)
model.add(Activation("tanh")) # [-1, +1]
model.add(Dense(1)) # Output Layer
model.add(Activation("sigmoid")) # [0, 1]
```

```python
sgd = SGD(lr=0.1)
model.compile(loss="binary_crossentropy", optimizer=sgd, metrics=["accuracy"])
model.fit(x, y, nb_epoch=200)
```

```python
preds = [model.predict(x_p) for x_p in x]
print(preds)
```
